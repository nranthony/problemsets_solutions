# [Cumulative Rainfall](https://www.practiceprobs.com/problemsets/python-numpy/expert/cumulative-rainfall/)

Your job is to monitor the rainfall ðŸŒ§ in a region that youâ€™ve broken into a 3x4 grid. Every time a storm comes, if a cell in the grid gets rain, you record it. At the end of a month you have the following two arrays of data.

```python
import numpy as np

rain_locations = np.array([
    [2,3],
    [0,1],
    [2,2],
    [2,3],
    [1,1],
    [2,3],
    [1,1]
])

rain_amounts = np.array([0.5,1.1,0.2,0.9,1.3,0.4,2.0])
```

Build a 3x4 array representing the total accumulated rainfall in each cell of the grid.

---

```python
rain_loc_amnts = np.zeros((len(rain_amounts),3,4))
for r in range(len(rain_amounts)):
    rain_loc_amnts[r,
                   rain_locations[r,0],
                   rain_locations[r,1]] = rain_amounts[r]
cumul_rain = rain_loc_amnts.sum(axis=0)
print(cumul_rain)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 8))
f = ax.imshow(cumul_rain, cmap='viridis')
ax.set_aspect('equal')
cbar = plt.colorbar(f, aspect=20, shrink=0.76)
cbar.set_label("Cumulative Rainfall (mm)", )
ax.xaxis.set_major_locator(plt.MaxNLocator(integer=True))
ax.yaxis.set_major_locator(plt.MaxNLocator(integer=True))
plt.show()
```
